```python
# eda_visuals.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load dataset
df = pd.read_csv('D:/AMITY/Semester_4/5. Major Project/mba-semester4-major-project-qollabb/ProjectApp/data/preprocessed/cleaned_citizen_feedback.csv')
```

```python
# Show basic info
print("\n--- Dataset Info ---")
print(df.info())
```

```python
# Check for missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
df.head(5)
```

```python

# Basic descriptive stats
print("\n--- Descriptive Statistics ---")
df.describe(include='all')
```

```python
# Plot Category Distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='city', order=df['city'].value_counts().index)
plt.title('Feedback Count by City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['local_service_satisfaction'], bins=8, kde=True)
plt.title('local_service_satisfaction Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Word Cloud of Feedback Text (optional for fun)
text = ' '.join(df['local_service_suggestions'].dropna()).join(df['park_suggestions'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Words in Feedback')
plt.show()
```

```python
# 01_eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from wordcloud import WordCloud


# Basic Info
print("\n--- Dataset Info ---")
print(df.info())

# Missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())

# Descriptive statistics
print("\n--- Descriptive Statistics ---")
print(df.describe(include='all'))

# Sample records
print("\n--- Sample Records ---")
print(df.sample(5))

# Create output dir
os.makedirs("../data/exports/eda_plots", exist_ok=True)

# 1. Age group distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='age_group', order=df['age_group'].value_counts().index)
plt.title('Age Group Distribution')
plt.tight_layout()
plt.savefig("../data/exports/eda_plots/age_group_dist.png")
plt.show()

# 2. Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.tight_layout()
plt.savefig("../data/exports/eda_plots/gender_dist.png")
plt.show()

# 3. City distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='city', order=df['city'].value_counts().index)
plt.title('City-wise Feedback Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../data/exports/eda_plots/city_dist.png")
plt.show()

# 4. Service Satisfaction Distribution
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='local_service_satisfaction')
plt.title('Local Service Satisfaction Levels')
plt.tight_layout()
plt.savefig("../data/exports/eda_plots/local_service_satisfaction.png")
plt.show()


```

```python
# 5. WordCloud: all suggestion columns
# text_cols = ["transport_suggestions", "park_suggestions", "library_suggestions", "local_service_suggestions"]
# all_text = " ".join(df[col].fillna("") for col in text_cols)
text = ' '.join(df['park_suggestions'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud from Suggestions')
plt.tight_layout()
plt.savefig("../data/exports/eda_plots/wordcloud.png")
plt.show()

print("\nEDA completed. Visuals saved to /data/exports/eda_plots")
```
